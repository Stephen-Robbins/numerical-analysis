```python
import numpy as np
import matplotlib.pyplot as plt

class lebesguePlotEquidistant:
    """
    A class for plotting the Lebesgue function for equidistant points.
    """
    
    def __init__(self, N):
        """
        Initializes the class with a number of points `N`.
        """
        self.N = N
       
        
    def x_j(self, i):
        """
        Calculates the i-th equidistant point.
        """
        return (-1+2*i/self.N)
    
    def L_j(self, j, points, t):
        """
        Calculates the j-th Lagrange basis polynomial evaluated at `t`.
        """
        y=1
        for i in range(self.N+1):
            if (i!=j):
                y=y*(t-points[i])/(points[j]-points[i])
                
        return y

    def LebFunc(self, points, t):
        """
        Calculates the Lebesgue function evaluated at `t`.
        """
        y=0
        for i in range(self.N+1):
            y=y+abs(self.L_j( i, points, t))
        return y
    
    def plot(self):
        """
        Plots the Lebesgue function over a given range of `t` values.
        """
        X=[self.x_j(i) for i in range(self.N+1)]
        t=np.arange(-1, 1, 0.0001)
        
        a=np.log10(self.LebFunc(X, t))
        
        plt.plot(t, a, label="Lebesgue function")
        plt.xlabel("t")
        plt.ylabel("log10(Lebesgue function)")
        plt.title("Lebesgue function for equidistant points (N={})".format(self.N))
        plt.legend()
        plt.show()


class lebesguePlotChebyshev:
    """
    A class for plotting the Lebesgue function for Chebyshev points.
    """
    
    def __init__(self, N):
        """
        Initializes the class with a number of points `N`.
        """
        self.N = N
       
        
    def x_j(self, i):
        """
        Calculates the i-th Chebyshev point.
        """
        return  np.cos(np.pi*(2*i+1)/(2*self.N+2))
    
    def L_j(self, j, points, t):
        """
        Calculates the j-th Lagrange basis polynomial evaluated at `t`.
        """
        y=1
        for i in range(self.N+1):
            if (i!=j):
                y=y*(t-points[i])/(points[j]-points[i])
                
        return y

    def LebFunc(self, points, t):
        """
        Calculates the Lebesgue function evaluated at `t`.
        """
        y=0
        for i in range(self.N+1):
            y=y+abs(self.L_j( i, points, t))
        return y
    
    def plot(self):
        """
        Plots the Lebesgue function over a given range of `t` values.
        """
        X=[self.x_j(i) for i in range(self.N+1)]
        t=np.arange(-1, 1, 0.0001)
        
        a=self.LebFunc(X, t)
        
        plt.plot(t, a, label="Lebesgue function")
        plt.xlabel("t")
        plt.ylabel("Lebesgue function")
        plt.title("Lebesgue function for Chebyshev points (N={})".format(self.N))
        plt.legend()
        plt.show()

        
```

```python
for n in [5, 10, 20]:
    lebesguePlotEquidistant(n).plot()
    
for n in [5, 10, 20]:
    lebesguePlotChebyshev(n).plot()
```
