```python
import matplotlib.pyplot as plt
import numpy as np

def Separable_Energy(N,R,x0):
    ''' 
        We use the Symplectic Integration Algorithm (SIA) with alpha=1, 
    Beta=1/2 for the separable Hamitonian: H=T(x_1)+V(x_2) where T(x)=V(x)= x^2/2
    in [0,R] and calculate and plot the energy.
    '''

    h=R/N
    x_1=[0 for i in range(N)]
    x_2=[0 for i in range(N)]
    x_1[0]=x0[0]
    x_2[0]=x0[1]
    for i in range(1, N):
        x_1[i]=x_1[i-1]-h*x_2[i-1]
        x_2[i]=x_2[i-1]+h*x_1[i]

    E=[x_1[i]**2/2+x_2[i]**2/2 for i in range(N)] 
    x=np.linspace(0, R, N)
    E_2= [x_1[0]**2/2+x_2[0]**2/2 for i in range(N)]

    E_3=[x_1[i]**2/2+x_2[i]**2/2 -h/2*x_1[i]*x_2[i] for i in range(N)] 

    plt.plot(x, E)
    plt.plot(x, E_2)
    plt.plot(x, E_3)
    plt.title(r"Initial conditions = {}".format(x0))
    plt.show()
```

```python
Separable_Energy(N=1000,R=20, x0=[0, .5])
Separable_Energy(N=1000,R=20, x0=[1, 4])
```
